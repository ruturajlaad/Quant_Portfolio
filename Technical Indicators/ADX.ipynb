```python
import pandas as pd
import datetime as dt
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
tickers=["ITC.NS"]
ohlcv_data={}
```

```python
for ticker in tickers:
    temp=yf.download(ticker,period='1mo',interval='5m')
    temp.dropna(how='any',inplace=True)
    ohlcv_data[ticker]=temp
```

```python
def ATR(DF,n=14):
    df=DF.copy()
    print(df)
    df["H-L"]=df["High"]-df["Low"]
    df["H-PC"]=df["High"]-df["Adj Close"].shift(1)
    df["L-PC"]=df["Low"]-df["Adj Close"].shift(1)
    df["TR"]=df[["H-L","H-PC","L-PC"]].max(axis=1,skipna=False)
    df["ATR"]=df["TR"].ewm(com=n,min_periods=n).mean()
    return df["ATR"]
```

```python
def ADX(DF,n=20):
    df=DF.copy()
    df["ATR"]=ATR(df,n)
    df["UpM"]=df["High"]-df["High"].shift(1)
    df["DwnM"]=df["Low"].shift(1)-df["Low"]
    df["+DM"]=np.where((df["UpM"]>df["DwnM"])&(df["UpM"]>0),df["UpM"],0)
    df["-DM"]=np.where((df["DwnM"]>df["UpM"])&(df["DwnM"]>0),df["DwnM"],0)
    df["+DI"]=100*(df["+DM"]/df["ATR"]).ewm(span=n,min_periods=n).mean()
    df["-DI"]=100*(df["-DM"]/df["ATR"]).ewm(span=n,min_periods=n).mean()
    df["ADX"]=abs((df["+DI"]-df["-DI"])/(df["+DI"]+df["-DI"])).ewm(span=n,min_periods=n).mean()
    return df["ADX"]
```

```python
for ticker in ohlcv_data:
    ohlcv_data[ticker]['ADX']=ADX(ohlcv_data[ticker])
```
